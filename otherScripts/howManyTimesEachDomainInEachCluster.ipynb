```python
inputFile = open ("everycdspfamdomains.txt","r")
outputFile = open ("howManyTimesEachDomainInEachCluster.txt","w")
clusterDictionary={}
currentCluster=""
```

```python
##The flag is useful so we know if we are looking for a new cluster or if we are still in the same one
##Each cluster finishes with "//" and we turn the flag on
##As we are not looking for scores in this script, we just add 1 to the number of times that domain appears for
##that cluster when we see ">>" followed by the name of the domain.
flag = True
counter=0
for line in inputFile:
    if flag:
        if line[:6]=="Query:":
            line=" ".join(line.split())
            line=line.split("|")
            currentCluster=line[0][7:]
            if currentCluster not in clusterDictionary.keys():
                clusterDictionary[currentCluster]={}
            flag = False
    else:
        if line[:2]=="//":
            flag=True
        elif line[:3]==">> ":
            domainName=line[3:].split(" ")[0]
            if domainName not in clusterDictionary[currentCluster].keys():
                clusterDictionary[currentCluster][domainName]=1
            else:
                clusterDictionary[currentCluster][domainName] += 1
```

```python
print (clusterDictionary)
```

```python
##Writes the whole dictionary to a file. One cluster per line
for geneClusterName in clusterDictionary.keys():
    line=geneClusterName+": "
    for domainName in sorted(clusterDictionary[geneClusterName].keys()):
        line+= domainName + ": " + str(clusterDictionary[geneClusterName][domainName]) +"; "
    outputFile.write(line+"\n")
```

```python
inputFile.close()
outputFile.close()
```
