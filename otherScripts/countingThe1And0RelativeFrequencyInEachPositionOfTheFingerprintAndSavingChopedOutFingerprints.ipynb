```python
##This scripts chops out all the positions of the fingerprints in which more than 99% of the times there is a 1 or a 0.
##It was intended at reducing the noise but these fingerprints got slightly worse results when using the machine learning algorithms.
```

```python
##Counts the percentage of 1 in each position of the fingerprint
inputFile = open ("fingerPrintsForIndividualClustersNOT478.txt")
position1Count={}
totalLines=0
for line in inputFile:
    i=0
    line=line.strip("\n")
    line= line[11:]
    for position in line:
        if position == "1":
            if i not in position1Count:
                position1Count[i]=1
            else:
                position1Count[i]+=1
        i+=1
    totalLines+=1
for i in range (0,306):
    if i not in position1Count:
        position1Count[i]=0
inputFile.close()
```

```python
for key in position1Count.keys():##Converts to percentage
    position1Count[key]/=totalLines
    position1Count[key]*=100
```

```python
for key in sorted(position1Count.keys()):
    print ("In the position %s, there is a %.2f %% of 1" % (key, position1Count[key] ))
```

```python
totalNumber=0
positionsToBeChopedOut=[]
for key in sorted(position1Count.keys()):
    if position1Count[key] > 99 or position1Count[key] <1:
        #print (key)
        positionsToBeChopedOut.append(key)
        totalNumber+=1
print ("There are %d positions to be excluded in the fingerprint" % totalNumber)
```

```python
##Counts the percentage of 1 in each position of the fingerprint
inputFile = open ("fingerPrintsForIndividualClustersNOT478.txt")
outputFile = open ("fingerprintsOnlyWithImportantSubstructures.txt","w")
for line in inputFile:
    i=0
    line=line.strip("\n")
    clusterName=line[0:11]
    outputFile.write(clusterName)
    line= line[11:]
    for position in line:
        if i not in positionsToBeChopedOut:
            outputFile.write(position)
        i +=1
    outputFile.write("\n")
inputFile.close()
outputFile.close()
```
